```python
copd_data['MWT1'].mean()

copd_data['MWT2'].mean()

copd_data['MWT1Best'].mean()

copd_data.fillna(copd_data['MWT1'].mean(), inplace=True)

copd_data.fillna(copd_data['MWT2'].mean(), inplace=True)

copd_data.fillna(copd_data['MWT1Best'].mean(), inplace=True)

copd_data.isnull().sum()

copd_data['COPDSEVERITY'].value_counts()
```

```python
#convert categorical columns to numerical values
copd_data.replace({'COPDSEVERITY':{'MODERATE':0,'SEVERE':1,'MILD':2,'VERY SEVERE':3}},inplace=True)

copd_data.head()

x = copd_data.drop(columns=['COPDSEVERITY'],axis=1)
y= copd_data['COPDSEVERITY']

x.head()

y.head()
```

```python
Model Training

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=0)


# Using Decision tree
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier().fit(x_train,y_train) # convert y_train to integer type
print('Accuracy of Decision Tree Classifier on training set : {:.2f}'.format(dtc.score(x_train,y_train)))
print('\t')
print('Accuracy of Decision Tree Classifier on test set : {:.2f}'.format(dtc.score(x_test,y_test)))
```

```python
input_data=(1,58,77,60,120,120,120,1.21,36,2.4,98,25,8,69.55,4,3,1,2,1,0,0,1,0)
# changing the input data into a numpy array
input_data_as_numpy_array=np.array(input_data)

# reshape the array
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=dtc.predict(input_data_reshaped)
print(prediction)


if prediction >= 0.0 and prediction < 1.0:
        print("MODERATE")
elif prediction >= 1.0 and prediction < 1.8:
        print("SEVERE")
elif prediction >= 1.8 and prediction < 2.8:
        print("MILD")
else:
        print("VERY SEVERE")
```

```python
from sklearn.tree import export_graphviz
from graphviz import Source
import os

tree_clf = DecisionTreeClassifier(criterion='entropy', max_depth=2)
tree_clf.fit(x, y)


feature_names = list(x.columns)
export_graphviz(
    tree_clf,
    out_file="copd_tree.dot",
    feature_names=feature_names,
    class_names=['MODERATE', 'SEVERE', 'MILD', 'VERY SEVERE'],
    rounded=True,
    filled=True
)

# Read and display the decision tree using graphviz.Source
with open("copd_tree.dot") as f:
    dot_graph = f.read()

Source(dot_graph)
```
